```python
from crewai import Agent, Task, Crew
```

```python
from langchain_ollama import ChatOllama
import os
```

```python
os.environ["OPENAI_API_KEY"] = "NA"

llm = ChatOllama(
    model="llama3.1",
    base_url="http://localhost:11434")
```

```python
from crewai_tools import FileReadTool
file_read_tool = FileReadTool(file_path='document.json')
```

```python
curator = Agent(
    role='Curator',
    goal='Review and Curate the question and summary response generated from the data',
    backstory=(
        "You are a Curator of the question and summary generated by external system. "
        "Your expertise lies in analysing given data file and understand its contents. "
        "You will indentify the data and qna parts from the given file and review the questions and answer from qna."
        "You have the ability to approve and reject the data contents"
        "You have ability to curate the questions and answer after reviewing the data contents"
    ),
    verbose=False,
    allow_delegation=False,
    tools=[file_read_tool],
    llm=llm
)
```

```python
task = Task(description=("Read the documents.json file. Identify the data content and qna content. "
                         "Analyse the data content. Review the questions and answer for the given content. "
                         "Generate alternate questions and summary for the content if qna is not appropriate for the data"),
            agent=curator,
            expected_output="text")

crew = Crew(
    agents=[curator],
    tasks=[task],
    verbose=True
)
result = crew.kickoff()

print(result)
```
